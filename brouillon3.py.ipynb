```python
def date_clean(x): 
    try: 

        months = {"January":1, "February":2, "March":3, "April":4, "May":5, "Jun":6, "July":7, "August":8, "September":9, "October":10, "November":11, "December":12}

        for month, numero in months.items(): 
            if month in x: 
                m = numero

        a = x.split(",")
        dia = a[0].split()[1]
        año = a[1].split()[0]

        return '{}-{}-{}'.format(dia, m, año)
    except: 
        return None
       
```
